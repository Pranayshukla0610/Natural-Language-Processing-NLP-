```python
import pandas as pd
import numpy as np
```

```python
data = pd.DataFrame({'text':['people watch Netflix','Netflix watch Netflix','people write reviews','Netflix write reviews'],'output':[1,1,0,0]})
data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
```

```python
bow = cv.fit_transform(data['text'])
bow
```

```python
print(cv.vocabulary_)
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
print(bow[2].toarray())
print(bow[3].toarray())
```

```python
cv.transform(["Netflix watch and write reviews of Netflix"]).toarray()
```
