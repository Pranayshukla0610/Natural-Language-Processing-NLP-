```python
import re
```

```python
def clean_resume(txt):
    cleantxt = re.sub('http\S+\s',' ',txt)
    cleantxt = re.sub('@\S+',' ',cleantxt)
    cleantxt = re.sub('#\S+',' ',cleantxt)
    cleantxt = re.sub('[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""), ' ', cleantxt)
    cleantxt = re.sub(r'[^\x00-\x7f]', ' ', cleantxt)
    cleantxt = re.sub('\s+',' ',cleantxt)
    return cleantxt
```

```python
clean_resume('my website #### $ #PranayShukla link is this http://helloworld and access it')
```

```python
data['Resume'] = data['Resume'].apply(lambda x: clean_resume(x))
```

```python
data['Resume'] = pd.DataFrame(data['Resume'])
```

```python
# Vectorization
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')
```

```python
tfidf.fit(data['Resume'])
requiredtext = tfidf.transform(data['Resume'])
```

```python
data
```

```python
requiredtext
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(requiredtext, data['Resume'], test_size=0.2, random_state=42)
```

```python
X_train.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score
```

```python
clf = OneVsRestClassifier(KNeighborsClassifier())
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
print(accuracy_score(y_test, y_pred))
```
