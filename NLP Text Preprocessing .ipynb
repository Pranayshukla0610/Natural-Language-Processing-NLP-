```python
#Stemming and Lemmatization

text = """Embracing life's challenges is like navigating a journey. ğŸš€
Stay motivated, overcome hurdles, and explore new paths to success!
Check out inspiring stories at https://motivationalhub.com for an extra boost!"""
```

```python
print(text)
```

```python
word_tokens = text.lower().split()
print(word_tokens)
```

```python
stemmer = nltk.stem.PorterStemmer()
stemmed_tokens = [stemmer.stem(token) for token in word_tokens]
print(stemmed_tokens)
```

```python
#Create Lemmatizer Object

lemmatizer = nltk.stem.WordNetLemmatizer()
lemmatized_token = [lemmatizer.lemmatize(token) for token in word_tokens]
print(lemmatized_token)
```

```python
#POS tagging in Lemmatization

def pos_tagger(nltk_tag):
    if nltk_tag.startswith("J"):
        return wordnet.ADJ
    elif nltk_tag.startswith("V"):
        return wordnet.ADJ
    elif nltk_tag.startswith("N"):
        return wordnet.NOUN
    elif nltk_tag.startswith("R"):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
tagged = nltk.pos_tag(word_tokens)
```

```python
print(tagged)
```

```python
# Get the root word for each of the tokens using their corresponding pos-tags
lemma_sent = []
for word, tag in tagged:
    new_tag = pos_tagger(tag)
    lemma = lemmatizer.lemmatize(word, new_tag)
    lemma_sent.append(lemma)
```

```python
print(f"Original sentence : \n{text}")
```

```python
print(f"Lemmatized sentence : \n{' '.join(lemma_sent)}")
```

```python
#NOISE ENTITY REMOVAL

text = """Embracing life's challenges is like navigating a journey. ğŸš€
Stay motivated, overcome hurdles, and explore new paths to success!
Check out inspiring stories at https://motivationalhub.com for an extra boost!"""

print(text)
```

```python
# Tokenization
word_tokens = text.lower().split()
```

```python
# PoS tagging
tagged = nltk.pos_tag(word_tokens)
```

```python
# Lemmatization
lemma_sent = []
for word, tag in tagged:
    new_tag = pos_tagger(tag)
    lemma = lemmatizer.lemmatize(word, new_tag)
    lemma_sent.append(lemma)
```

```python
#REMOVING STOPWORDS

# Obtain the list of stopwords from the corpus
stp_wrds_eng = stopwords.words('english')
print(stp_wrds_eng)
```

```python
# Removing stopwords
text_clean = [w for w in lemma_sent if w not in stp_wrds_eng]
print(f"Lemmatized : \n{' '.join(lemma_sent)}")
print(f"Cleaned  : \n{' '.join(text_clean)}")
```

```python
#REMOVING URLs
# Identifying and substituting urls using the pattern 'https\S+' for urls
text_clean = re.sub(r'http\S+', '', ' '.join(text_clean), flags=re.MULTILINE)
print(text_clean)
```

```python
#Remove Punctuations

text_clean = [w for w in text_clean if w not in punctuation]
print(f"Cleaned  : \n{''.join(text_clean)}")
```

```python
#REMOVE EMOTICONS

RE_EMOJI = re.compile('[\U00010000-\U0010ffff]', flags=re.UNICODE)
def strip_emoji(text):
    return RE_EMOJI.sub(r'', text)
```

```python
# Use function to remove emoticons from text
text_clean = strip_emoji(''.join(text_clean))
print(text_clean)
```

```python
nltk.download('product_reviews_1')
nltk.download('punkt')
nltk.download('stopwords')
```

```python
file_ids = nltk.corpus.product_reviews_1.fileids()
print(file_ids)
```

```python
reviews = nltk.corpus.product_reviews_1.reviews(file_ids[1])
raw_text = nltk.corpus.product_reviews_1.raw(file_ids[1])
review_sentences = nltk.corpus.product_reviews_1.sents(file_ids[1])
```

```python
review_titles = {f"rev_{i+1}" : reviews[i].title for i in range (len(reviews))}
for tit in review_titles.keys():
    print(f"{tit}:{review_titles[tit]}")
```

```python
review_texts = []
for i in range(len(reviews)):
    review_lines = reviews[i].review_lines
    line = ""
    for j in range (len(review_lines)):
        line += ' '.join(review_lines[j].sent)
    review_texts.append(line)
```

```python
print("Complete Review Texts".center(150))
i = 1
for rev in review_texts:
    print(f"Title --> {review_titles[f'rev_{i}']} :\n\n{rev}\n")
    print(f"Summary --> \nLen of review : {len(rev.split())} words.\n")
    i += 1
```

```python
words_per_review = [len(rev) for rev in review_texts]
```

```python
print(f"Largest review: {max(words_per_review)}")
print(f"Smallest review: {min(words_per_review)}")
print(f"Average words per review: {sum(words_per_review)/ len(words_per_review):.0f}")
```

```python
sent_length = [len(rev) for rev in review_sentences]
```

```python
print(f"Largest sentence: {max(sent_length)}")
print(f"Smallest sentence: {min(sent_length)}")
print(f"Average sentence length: {sum(sent_length)/ len(sent_length):.0f}")
```
